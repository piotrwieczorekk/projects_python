<b><h2>Webscraping data about books in swedish with BeautifulSoup</h2></b>

### Libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import requests
```

### Connection

```python
url = 'https://www.bokus.com/cgi-bin/product_search.cgi?language=Svenska'
response = requests.get(url)
response.status_code
```

### Scraping code - get data about swedish books across 50 pages

```python
soup = BeautifulSoup(response.content,'html.parser')
swedish_book = soup.find_all('div',class_='Item__info')
```

```python
k = []
for page in range(1,51):
    url = f'https://www.bokus.com/cgi-bin/product_search.cgi?is_paginate=1&language=Svenska&from={page}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content,'html.parser')
    swedish_book = soup.find_all('div',class_='Item__info')
    for book in swedish_book:
        title = book.find('h3',class_='Item__title Item__title--large').a.text.strip()
        book_link = 'https://www.bokus.com' + book.find('h3',class_='Item__title Item__title--large').a.get('href')
        try:
            author = book.find('h4',class_='Item__authors ProductList__authors').text.strip()
        except:
            author = np.NaN
        try:
            book_type = book.find('span',class_='Item__format-as-link').text.strip()
        except:
            book_type = np.NaN
        try:
            recommended_age = book.find('div',class_='Item__tags').a.text.strip()
        except:
            recommended_age = np.NaN
        actual_price = book.find('div',class_='pricing pricing--large ProductList__pricing u-m-t--1').find_all('span')[0].text.strip()
        try:
            old_price = book.find('div',class_='pricing pricing--large ProductList__pricing u-m-t--1').find_all('span')[1].text.strip()
        except:
            old_price = np.NaN
        book_info = book.find('div',class_='Item__edition').text.strip()
        k.append({'title':title,
              'book_link':book_link,
              'author':author,
              'book_type':book_type,
              'book_info':book_info,
              'recommended_age':recommended_age,
              'actual_price':actual_price,
              'old_price':old_price})
    time.sleep(np.random.randint(2,4))
    print(f'Getting page {page}. Waiting...')
```

### Data cleaning

```python
df = pd.DataFrame(k)
```

```python
df.head(10)
```

```python
df['title']
```

```python
df['title'] = df['title'].apply(lambda x: x.split(' \n',1)[0])
```

```python
df['author']
```

```python
df['author'] = df['author'].apply(lambda x: x.replace('av\n ','').translate(str.maketrans({'\n':''})) if type(x) == str else x)
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'].apply(lambda x: x.split(': ')[1] if ': ' in x else x))
```

```python
df['old_price'] = pd.to_numeric(df['old_price'])
```

```python
df.rename(columns={'actual_price':'actual_price(SEK)',
                   'old_price':'old_price(SEK)'},inplace=True)
```
