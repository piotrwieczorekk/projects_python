<h2><b>Scraping data about Serie A statistics (season 2022/2023)</h2></b>

### Libraries

```python
import pandas as pd
import numpy as np
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from IPython.display import Image
```

### Connection

```python
url = "https://www.legaseriea.it/en/team/napoli"
response = requests.get(url)
response.status_code
```

```python
PATH = 'C:\webdrivers\chromedriver.exe'
driver = webdriver.Chrome(executable_path=PATH)
url = "https://www.legaseriea.it/en/team/napoli"
driver.get(url)
driver.implicitly_wait(3)
cookies = driver.find_element(By.XPATH,'//button[@class="iubenda-cs-accept-btn iubenda-cs-btn-primary"]')
cookies.click()
view_table = driver.find_element(By.XPATH,'(//a[@class="p3 bold white"])[1]')
view_table.click()
driver.implicitly_wait(5)
links = []
statistics_links = []
statistics_data = []
team_points = []
```

```python
points = driver.find_elements(By.XPATH,'.//h3[@class="bold primary-200 text-center"]')
for h in points:
    team_points.append({'points':h.text.strip()})
```

```python
df_points = pd.DataFrame(team_points)
```

```python
df_points
```

### Getting links to all teams in the table

```python
team_links = driver.find_elements(By.XPATH,'.//table[@class="table hm-table hm-table-classify"]//tbody//tr//td[2]//a')
for i in team_links:
    team_link = i.get_attribute('href')
    links.append({"link":team_link})
```

```python
links
```

### Getting links to all pages with statistics for each team

```python
for k in links:
    driver.get(k["link"])
    driver.implicitly_wait(3)
    statistics_page = driver.find_element(By.XPATH,'.//div[@class="hm-nav-section"]//li[5]//a').get_attribute('href')
    statistics_links.append({"link_stats":statistics_page})
```

```python
statistics_links
```

```python
for u in statistics_links:
    driver.get(u["link_stats"])
    driver.implicitly_wait(3)
    team_name = driver.find_element(By.XPATH,'//h3[@class="MonumentExtended regular regular white"]').text.strip()
    wins = driver.find_element(By.XPATH,'(.//p[@class="black bold ms-auto"])[1]').text.strip()
    draws = driver.find_element(By.XPATH,'(.//p[@class="black bold ms-auto"])[2]').text.strip()
    loses = driver.find_element(By.XPATH,'(.//p[@class="black bold ms-auto"])[3]').text.strip()
    fouls_against = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-200"])[1]').text.strip()
    fouls = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-200"])[2]').text.strip()
    yellow_cards = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-200"])[3]').text.strip()
    red_cards = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-200"])[4]').text.strip()
    double_yellow_cards = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-200"])[5]').text.strip()
    total_goals_scored = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-green"])[1]').text.strip()
    big_chances_missed = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-green"])[2]').text.strip()
    assist_leading_to_goals = driver.find_element(By.XPATH,'(.//h5[@class="MonumentExtended regular uppercase primary-green"])[3]').text.strip()
    statistics_data.append({"team_name":team_name,
                            "wins":wins,
                            "draws":draws,
                            "loses":loses,
                            "fouls_against":fouls_against,
                            "fouls":fouls,
                            "yellow_cards":yellow_cards,
                            "red_cards":red_cards,
                            "double_yellow_cards":double_yellow_cards,
                            "red_cards":red_cards,
                            "double_yellow_cards":double_yellow_cards,
                            "total_goals_scored":total_goals_scored,
                            "big_chances_missed":big_chances_missed,
                            "assist_leading_to_goals":assist_leading_to_goals})
    time.sleep(3)
```

### Final data frame

```python
df_stats = pd.DataFrame(statistics_data)
```

```python
df_stats
```

```python
df_table = pd.merge(df_stats, df_points, left_index=True, right_index=True)
```
