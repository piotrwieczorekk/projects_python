<h2><b>Webscraping data about french podcast episodes with Selenium</b></h2>

### Libraries

```python
import pandas as pd
import numpy as np
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
```

### We are going to iterate 5 times across multiple pages with french podcasts, get all needed podcast links and then open each of them to get needed data

```python
PATH = 'C:\webdrivers\chromedriver.exe'
driver = webdriver.Chrome(executable_path=PATH)
all_links = []
episode_info = []
for page in range(1,6):
    page_url = f"https://www.easyfrench.fm/page/{page}"
    driver.get(page_url)
    driver.implicitly_wait(3)
    podcast_links = driver.find_elements(By.XPATH,".//h3//a")
    for i in podcast_links:
        link = i.get_attribute('href')
        all_links.append({"link":link})
    time.sleep(2)
for k in all_links:
    driver.get(k["link"])
    episode_link = k["link"]
    try:
        episode_number = driver.find_element(By.XPATH,'//div[@class="hero-info"]/h5').text.strip()
    except:
        episode_number = np.NaN
    episode_title = driver.find_element(By.XPATH,'//div[@class="hero-info"]/h1').text.strip()
    episode_date = driver.find_element(By.XPATH,'(//div[@class="column"]/p)[1]').text.strip()
    about_episode = driver.find_element(By.XPATH,'(//div[@class="split-primary prose"]/p)[1]').text.strip()
    podcast_apps = driver.find_elements(By.XPATH,'.//nav[@class="links link-list-inline-row"]//ul//li//a')
    result = [category.text for category in podcast_apps]
    episode_info.append({'episode_link':episode_link,
                         'episode_number':episode_number,
                         'episode_title':episode_title,
                         'episode_date':episode_date,
                         'about_episode':about_episode,
                         'podcast_apps':podcast_apps,
                         'podcast_apps':result[1:9]})
    time.sleep(2)
```

### Final data frame

```python
pd.DataFrame(episode_info)
```
