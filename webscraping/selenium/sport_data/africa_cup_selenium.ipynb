## **Scraping data about Africa Cup of Nations from the site: https://www.sofascore.com/tournament/football/africa/africa-cup-of-nations/270**

### Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import seaborn as sns
from selenium.webdriver.support import expected_conditions as EC
import matplotlib.pyplot as plt
```

### The code

```python
url = "https://www.sofascore.com/tournament/football/africa/africa-cup-of-nations/270"
group_data = []
s = Service('C:\webdrivers\chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get(url)

WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '(//p[@class="fc-button-label"])[1]')))
cookies = driver.find_element(By.XPATH,'(//p[@class="fc-button-label"])[1]')
cookies.click()
driver.refresh()
WebDriverWait(driver, 45).until(EC.presence_of_element_located((By.XPATH, '(//div[@class="sc-jlZhew kgghii"]//div[@style="width: 100%;"])[1]')))


for i in range(1,7):  
    group_tables = driver.find_elements(By.XPATH,f'(//div[@class="sc-jlZhew kgghii"]//div[@style="width: 100%;"])[{i}]')
    for group_table in group_tables:
        table_rows = group_table.find_elements(By.XPATH,'.//div[@class="sc-fqkvVR sc-dcJsrY djvcFd fFmCDf sc-11dda344-0 bsFunH"]')
        for each_row in table_rows:
            team_position = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR byYarT"][1]').text.strip()
            team_name = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR fQCbLG"]').text.strip()
            matches_played = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR byYarT"][3]').text.strip()
            wins = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR byYarT"][4]').text.strip()
            draws = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR byYarT"][5]').text.strip()
            losts = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR byYarT"][6]').text.strip()
            goals = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR byYarT"][7]').text.strip()
            points = each_row.find_element(By.XPATH,'.//div[@class="sc-fqkvVR byYarT"][8]').text.strip()
            group_data.append({"team_pos":team_position,
                               "team_name":team_name,
                               "matches_played":matches_played,
                               "wins":wins,
                               "draws":draws,
                               "losts":losts,
                               "goals":goals,
                               "points":points})
        time.sleep(2)
    time.sleep(2)

```

```python
df = pd.DataFrame(group_data)
```

### Tournament groups A-F

```python
group_A = df.iloc[0:4,:]
group_B = df.iloc[4:8,:]
group_C = df.iloc[8:12,:]
group_D = df.iloc[12:16,:]
group_E = df.iloc[16:20,:]
group_F = df.iloc[20:24,:]
```
