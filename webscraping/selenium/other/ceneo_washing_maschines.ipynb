## **Scraping data about washing maschines from ceneo.pl**

### Libraries

```python
import requests
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import seaborn as sns
from selenium.webdriver.support import expected_conditions as EC
import matplotlib.pyplot as plt
```

### Scraping links to products

```python
url = "https://www.ceneo.pl/Pralki"
washing_mashine_links = []
s = Service('C:\webdrivers\chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get(url)

WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, '//button[@class="cookie-consent__buttons__action js_cookie-consent-agree primary"]')))
cookies = driver.find_element(By.XPATH,'//button[@class="cookie-consent__buttons__action js_cookie-consent-agree primary"]')
cookies.click()
driver.refresh()

for page in range(1,38):
    scraped_links = driver.find_elements(By.XPATH,'//strong[@class="cat-prod-row__name"]//a[1]')
    for scraped_link in scraped_links:
        scraped_link = scraped_link.get_attribute('href')
        washing_mashine_links.append(scraped_link)
    if page != 37:
        next_page = driver.find_element(By.XPATH,'//a[@class="pagination__item pagination__next"]')
        next_page.click()
        print(f'Scraping product links from page {page}...')
    time.sleep(5)
```

```python
washing_mashine_links = list(filter(lambda url: "Offer" not in url, washing_mashine_links))
```

```python
washing_mashine_links[0:5]
```

### Scraping data about 50 washing maschines

```python
all_data = []
for each_url in washing_mashine_links[0:50]:
    driver.get(each_url)
    try:
        desc = driver.find_element(By.XPATH,'//div[@class="product-top__product-info__tags"]').text.strip()
    except:
        desc = np.nan
    try:
        name = driver.find_element(By.XPATH,'//div[@class="product-top__title"]').text.strip()
        print(f'Getting data about {name}...')
    except:
        name = np.nan
    try:
        score = driver.find_element(By.XPATH,'//span[@class="product-review__score"]').text.strip()
    except:
        score = np.nan
    try:
        number_reviews = driver.find_element(By.XPATH,'//span[@class="product-review__qo"]').text.strip()
    except:
        number_reviews = np.nan
    try:
        lowest_price = driver.find_element(By.XPATH,'//div[@class="product-offer-summary"]//span[@class="value"]').text.strip()
    except:
        lowest_price = np.nan
    try:
        free_delivery = driver.find_element(By.XPATH,'//div[@class="free-delivery-label"]').text.strip()
    except:
        free_delivery = np.nan
    all_data.append({"desc":desc,
                     "name":name,
                     "score":score,
                     "num_reviews":number_reviews,
                     "lowest_price":lowest_price,
                     "free_delivery":free_delivery})
    time.sleep(2)
```
