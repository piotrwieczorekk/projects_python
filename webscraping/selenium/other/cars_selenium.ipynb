## **Scraping data about cars (polish market)**

## Libraries

```python
import requests
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import seaborn as sns
from selenium.webdriver.support import expected_conditions as EC
import matplotlib.pyplot as plt
```

## Scraping code

```python
url = "https://www.otomoto.pl/osobowe"
s = Service('C:\webdrivers\chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get(url)

WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//div[@id="onetrust-button-group"]//button[@id="onetrust-accept-btn-handler"]')))

cookies = driver.find_element(By.XPATH,'//div[@id="onetrust-button-group"]//button[@id="onetrust-accept-btn-handler"]')
cookies.click()
driver.refresh()

car_data = []

for page in range(1,101):
    url = f"https://www.otomoto.pl/osobowe?page={page}"
    driver.get(url)
    print(f'Getting data from page number {page}')
    all_articles = driver.find_elements(By.XPATH,'//article[@class="ooa-yca59n efpuxbr0"]')
    for each_article in all_articles:
        try:
            car_info = each_article.find_element(By.XPATH,'.//p[@class="efpuxbr10 ooa-1tku07r er34gjf0"]').text.strip()
        except:
            car_info = np.nan
        try:
            car_fuel = each_article.find_element(By.XPATH,'.//dd[@class="ooa-1omlbtp efpuxbr13"][2]').text.strip()
        except:
            car_fuel = np.nan
        try:
            car_driven_km = each_article.find_element(By.XPATH,'.//dd[@class="ooa-1omlbtp efpuxbr13"][1]').text.strip()
        except:
            car_driven_km = np.nan
        try:
            car_gearbox = each_article.find_element(By.XPATH,'.//dd[@class="ooa-1omlbtp efpuxbr13"][3]').text.strip()
        except:
            car_gearbox = np.nan
        try:
            car_year =  each_article.find_element(By.XPATH,'.//dd[@class="ooa-1omlbtp efpuxbr13"][4]').text.strip()
        except:
            car_year = np.nan
        try:
            car_price = each_article.find_element(By.XPATH,'.//h3[@class="efpuxbr16 ooa-1n2paoq er34gjf0"]').text.strip()
        except:
            car_price = np.nan
        car_data.append({"info":car_info,
                         "fuel":car_fuel,
                         "driven_km":car_driven_km,
                         "gearbox_type":car_gearbox,
                         "year":car_year,
                         "price":car_price})
    time.sleep(1)
```

## Cleaning the data

```python
df = pd.DataFrame(car_data)
```

```python
df
```

```python
df['gearbox_type'].unique()
```

```python
df['gearbox_type'].replace({'Automatyczna':"Automatic",
                            "Manualna":"Manual"},inplace=True)
```

```python
df[['driven_km','price']] = df[['driven_km','price']].map(lambda x: x.replace(' ','').replace('km','') if isinstance(x,str) else x)
```

```python
df['info']
```

```python
additional_data = []

for i in df['info'].apply(lambda x: x.split(' • ') if isinstance(x,str) else x):
    try:
        engine_capacity = i[0]
    except:
        engine_capacity = np.nan
    try:
        engine_power = i[1]
    except:
        engine_power = np.nan
    try:
        car_desc = i[2]
    except:
        car_desc = np.nan
    additional_data.append({"engine_capacity":engine_capacity,
                            "engine_power":engine_power,
                            "car_desc":car_desc})
```

```python
df
```

```python
pd.DataFrame(additional_data)
```

```python
df = pd.concat([df,pd.DataFrame(additional_data)],axis=1).iloc[:,1:]
```

```python
df['fuel'].replace({"Benzyna":"Petrol",
                    "Benzyna+LPG":"Petrol+LPG",
                    "Elektryczny":"Electric",
                    "Hybryda Plug-in":"Plug-in Hybrid"},
                    inplace=True)
```
