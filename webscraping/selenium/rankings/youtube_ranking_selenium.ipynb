<h2><b>Scraping data about most popular youtube videos</h2></b>

<h3>Libraries</h3>

```python
import pandas as pd
import numpy as np
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from IPython.display import Image
```

<h3>Status code and connection</h3>

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
response = requests.get(url)
response.status_code
```

```python
PATH = 'C:\webdrivers\chromedriver.exe'
driver = webdriver.Chrome(executable_path=PATH)
driver.get(url)
driver.implicitly_wait(10)
```

Finding the table containing data about videos and all tr within the table

```python
yt_vid = driver.find_elements(By.XPATH,"(//table[@class='wikitable sortable jquery-tablesorter'])[1]/tbody/tr")
```

```python
Image(url="table1.png",width=1500, height=600)
```

<h3>Getting the headers</h3>

```python
headers = [i.text.strip() for i in driver.find_elements(By.XPATH,"(//table[@class='wikitable sortable jquery-tablesorter'])[1]/thead/tr/th")]
```

```python
headers = headers[0:5]
```

```python
headers
```

<h3>Scraping code</h3>

```python
v = []
for vid in yt_vid:
    rank = vid.find_element(By.XPATH,".//td[1]").text.strip()
    vid_name = vid.find_element(By.XPATH,".//td[2]").text.strip()
    uploader  = vid.find_element(By.XPATH,".//td[3]").text.strip()
    views_billions = vid.find_element(By.XPATH,".//td[4]").text.strip()
    pub_date = vid.find_element(By.XPATH,".//td[5]").text.strip()
    v.append({headers[0]:rank,
              headers[1]:vid_name,
              headers[2]:uploader,
              headers[3]:views_billions,
              headers[4]:pub_date})
```

<h3>Data cleaning</h3>

```python
yt_df = pd.DataFrame(v)
```

```python
yt_df
```

```python
yt_df.dtypes
```

```python
yt_df['Video name'] = yt_df['Video name'].apply(lambda x: x.split('[')[0])
```

```python
yt_df['Video name'].head(10)
```

```python
yt_df['Views (billions)'] = yt_df['Views (billions)'].astype('float')
```

```python
yt_df.dtypes
```

```python
yt_df['Video name'] = yt_df['Video name'].apply(lambda x: x.replace('"',''))
```

```python
yt_df['Publication date'] = pd.to_datetime(yt_df['Publication date'])
```
